```python
import numpy as np
import matplotlib.pyplot as plt

a = list(range(0,361,20))
x = [i*(np.pi/180) for i in a]
plt.plot(a,np.sin(x))
plt.plot(a,np.cos(x))
plt.plot(a,np.tan(x))
plt.title("TRIGONOMETRIC FUNCTIONS")
plt.xlabel("Angle")
plt.ylabel("Value")
plt.show()
```
